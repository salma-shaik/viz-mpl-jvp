```python
%matplotlib inline
## embeds static images into the notebook

## either matplotlib inline or plt.show()


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure()

plt.plot(x, np.sin(x), linestyle='-')
plt.plot(x, np.cos(x), linestyle='--')

fig.savefig('fig/sin_cos.png')
```
